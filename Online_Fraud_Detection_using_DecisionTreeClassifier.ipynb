```python
import pandas as pd
import numpy as np
data = pd.read_csv("/content/Credit Card.csv")
print(data.head())
```

```python
data_cleaned = data.dropna()
```

```python
print(data_cleaned.isnull().sum())
```

```python
#Now, our data is clean
# Exploring transaction type
print(data_cleaned.type.value_counts())
```

```python
type = data_cleaned["type"].value_counts()
transactions = type.index
quantity = type.values

import plotly.express as px
figure = px.pie(data_cleaned,
             values=quantity,
             names=transactions,hole = 0.5,
             title="Distribution of Transaction type")
figure.show()
```

```python
#Now let’s have a look at the correlation between the features of the data with the isFraud column
# Checking correlation
correlation = data_cleaned.corr()
print(correlation["isFraud"].sort_values(ascending=False))
```

```python
#Now let’s transform the categorical features into numerical
data_cleaned["type"] = data_cleaned["type"].map({"CASH_OUT": 1, "PAYMENT": 2,
                                 "CASH_IN": 3, "TRANSFER": 4,
                                 "DEBIT": 5})
data_cleaned["isFraud"] = data_cleaned["isFraud"].map({0: "No Fraud", 1: "Fraud"})
print(data_cleaned.head())
```

```python
#Online Payments Fraud Detection Model
# splitting the data
from sklearn.model_selection import train_test_split
x = np.array(data_cleaned[["type", "amount", "oldbalanceOrg", "newbalanceOrig"]])
y = np.array(data_cleaned[["isFraud"]])
```

```python
# training a machine learning model
from sklearn.tree import DecisionTreeClassifier
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.10, random_state=42)
model = DecisionTreeClassifier()
model.fit(xtrain, ytrain)
print(model.score(xtest, ytest))
```

```python
# prediction
#features = [type, amount, oldbalanceOrg, newbalanceOrig]
features = np.array([[4, 9000.60, 9000.60, 0.0]])
print(model.predict(features))
```
